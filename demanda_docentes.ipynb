**Cálculo final de la demanda**

Finalmente, el requerimiento de horas en Secundaria, así como las secciones en los demás niveles, permitirá identificar el número de docentes.

Pero para dichos fines, primero estimamos la cantidad de horas docente que se necesitan por cada nivel. Recordemos que las horas totales en Inicial y Primaria son idénticas en EIB y no EIB.

De este modo, para el nivel Inicial EBR, Primaria EBR, Inicial e Intermedio EBA, Inicial EBE y Primaria EBE, el cálculo de la demanda toma la forma:

$$D_{l,i} = S_{l,i} \times m_{1,l}$$

Donde $m_1$ es el multiplicador de docentes por sección. Asimismo, para hallar la demanda de docentes para el nivel Secundaria EBR y Avanzado EBA, se presenta la siguiente ecuación:

$$D_{l,i,c} = S_{l,i,c} \times m_{2,l,c} \times m_{3,l}^{-1}$$

Donde $m_{2,l,c}$ es el multiplicador de horas por sección para cada especialidad de enseñanza $c$, y $m_{3,l}$ el multiplicador de horas por docente.

**Ejemplos**
1. Tenemos una escuela llamada "A", de nivel Inicial con 4 secciones. Como se muestra en un inicio, la cantidad de docentes por sección requerida en el nivel de Inicial es de 1 docente por sección, con esto tenemos que $S_{l,i}=S_{Inicial,A}=4$, y $m_{1,l}=m_{1,Inicial}=1$. Hallamos la demanda de docentes como: $D_{Inicial,A} = S_{Inicial,A} \times m_{1,Inicial} = 4 \times 1 = 4$; es decir, se requerirán 4 docentes en total.


2. Tenemos una escuela llamada "B", de nivel Secundaria, de Jornada Escolar Regular, con 20 secciones. En este caso, cada año de estudio (primero hasta quinto) cuenta con 4 secciones, haciendo un total de 20 secciones. Por otro lado, sabemos que el requerimiento de horas por cada especialidad es homogéneo en cada año de estudios, con lo cual, por ejemplo, para Matemática, se requieren 4 horas semanales por igual para cada año de estudios. Así, podemos resumir los datos de la siguiente manera:

Distribución de secciones para cada especialidad (son 20 secciones, 4 para cada año de estudios). Sin embargo, cada año de estudios tiene un mismo requerimiento de horas docente para cada especialidad de enseñanza.
$$S_{l,i,c} = S_{Secundaria,B,Comunicación} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Matemática} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Inglés} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Arte y Cultura} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Ciencias Sociales} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Ed Física} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Ed Religiosa} = 20$$
$$S_{l,i,c} = S_{Secundaria,B,Ciencia y Tecnología} 20$$
$$S_{l,i,c} = S_{Secundaria,B,Ed para el Trabajo} = 20$$

Por otro lado, la cantidad de horas por especialidad para Secundaria EBR, de acuerdo al CNEB 2016 es:
$$m_{2,l,c} = m_{2,Secundaria,Comunicación} = 4$$
$$m_{2,l,c} = m_{2,Secundaria,Matemática} = 4$$
$$m_{2,l,c} = m_{2,Secundaria,Inglés} = 3$$
$$m_{2,l,c} = m_{2,Secundaria,Arte y Cultura} = 3$$
$$m_{2,l,c} = m_{2,Secundaria,Ciencias Sociales} = 3$$
$$m_{2,l,c} = m_{2,Secundaria,Ed Física} = 3$$
$$m_{2,l,c} = m_{2,Secundaria,Ed Religiosa} = 2$$
$$m_{2,l,c} = m_{2,Secundaria,Ciencia y Tecnología} 4$$
$$m_{2,l,c} = m_{2,Secundaria,Ed para el Trabajo} = 2$$

Finalmente, la jornada de docentes destinada a aula es de un total de 24 a 26 horas semanales para EBR Secundaria JER. Asumimos por simplicidad que son 25 horas. Con ello, se tiene:
$$m_{3,l}=m_{3,Secundaria}=25$$

Reemplazando, se tiene la siguiente demanda de docentes por especialidad:

$$D_{l,i,c} = S_{l,i,c} \times m_{2,l,c} \times m_{3,l}^{-1}$$
$$D_{Secundaria,B,Comunicación} = 20 \times 4 \times 25^{-1} \equiv 4$$
$$D_{Secundaria,B,Matemática} = 20 \times 4 \times 25^{-1} \equiv 4$$
$$D_{Secundaria,B,Inglés} = 20 \times 3 \times 25^{-1} \equiv 3$$
$$D_{Secundaria,B,Arte y Cultura} = 20 \times 3 \times 25^{-1} \equiv 3$$
$$D_{Secundaria,B,Ciencias Sociales} = 20 \times 3 \times 25^{-1} \equiv 3$$
$$D_{Secundaria,B,Ed Física} = 20 \times 3 \times 25^{-1} \equiv 3$$
$$D_{Secundaria,B,Ed Religiosa} = 20 \times 2 \times 25^{-1} \equiv 2$$
$$D_{Secundaria,B,Ciencia y Tecnología} = 20 \times 4 \times 25^{-1} \equiv 4$$
$$D_{Secundaria,B,Ed para el Trabajo} = 20 \times 2 \times 25^{-1} \equiv 2$$
